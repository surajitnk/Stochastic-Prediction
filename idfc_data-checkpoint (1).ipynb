# Data Cleaning

## Handling Missing Values


```python
# filling NaN values of Max_Ratio_OC_Pending_POS,CoAp_Income,Total_Field_Trails,Total_Resolved with default value as 0
data['Max_Ratio_OC_Pending_POS'].fillna(0, inplace=True)
data['CoAp_Income'].fillna(0, inplace=True)
data['Total_Field_Trails'].fillna(0, inplace=True)
data['Total_Resolved'].fillna(0, inplace=True)
```

```python
# Calculate the mean of the Max_Perc_Def_Chg_Pending column
mean_value = data['Max_Perc_Def_Chg_Pending'].mean()

# Fill missing values in Max_Perc_Def_Chg_Pending with the mean
data['Max_Perc_Def_Chg_Pending'].fillna(mean_value, inplace=True)
```

```python
# Impute missing values in Perc_Paymode_Cheq_Fail with the mean
mean_value = data['Perc_Paymode_Cheq_Fail'].mean()
data['Perc_Paymode_Cheq_Fail'].fillna(mean_value, inplace=True)
```

```python
# Calculate the mean of the Max_Loan_Balance_Others column
mean_value = data['Max_Loan_Balance_Others'].mean()

# Fill missing values in Max_Loan_Balance_Others with the mean
data['Max_Loan_Balance_Others'].fillna(mean_value, inplace=True)

original_data= data
```

## Logistic Regression

```python
from sklearn.impute import SimpleImputer

# Initialize the imputer
imputer = SimpleImputer(strategy='mean')

# Fit and transform the imputer on your training data
train_X_imputed = imputer.fit_transform(train_X)

# Transform the validation data using the same imputer
val_X_imputed = imputer.transform(val_X)

# Now, you can use train_X_imputed and val_X_imputed in your Logistic Regression model.
```

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Assuming you have already split your data into train_X, train_y, val_X, val_y as mentioned earlier

# Initialize the Logistic Regression model
logistic_reg_model = LogisticRegression(class_weight='balanced', random_state=88)

# Fit the model on the training data
logistic_reg_model.fit(train_X_imputed, train_y.values.ravel())  # Using .values.ravel() to flatten train_y if needed

# Make predictions on the validation data
val_predictions = logistic_reg_model.predict(val_X_imputed)

# Evaluate the model's performance
accuracy = accuracy_score(val_y, val_predictions)
confusion_mat = confusion_matrix(val_y, val_predictions)
classification_rep = classification_report(val_y, val_predictions)

# Print the results
print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(confusion_mat)
print("Classification Report:")
print(classification_rep)
```

By setting class_weight='balanced', you've improved the model's performance on the minority class (class 1), as evidenced by an increase in recall for class 1. However, this comes at the cost of decreased precision for class 1, which results in lower F1-score for that class. It's a trade-off between precision and recall.

Here's an interpretation of the results:
